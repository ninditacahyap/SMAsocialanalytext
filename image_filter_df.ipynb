```python
import os
import glob
import re

ig_path = '/workspaces/sma-plan2/instagram'

def get_images_list(instagram):
    pattern = r'UTC_[1-9]\.jpg$'
    file_list = glob.glob(os.path.join(os.path.join(ig_path, instagram), '*.jpg'))
    result = [item for item in file_list if re.search(pattern, item)]
    return result
```

```python
def filter_images(instagram, df):
    file_list = get_images_list(instagram)
    file_list = [x.split('/')[-1] for x in file_list]
    file_list = [x[:19] for x in file_list]
    file_list = [f"{s.split('_')[0]} {s.split('_')[1].replace('-', ':')}" for s in file_list]
    file_list = list(set(file_list))

    temp_df = df.copy()
    temp_df = temp_df[temp_df["Date"].isin(file_list)]
    
    temp_df.to_csv()
```
